```python
from analysis_tools import DataLoader
```

### Open the file

```python
FILE = "/eos/experiment/wcte/data/2025_commissioning/processed_offline_data/production_v1_0/1928/WCTE_merged_production_R1928.root"
loader = DataLoader(FILE)
```

### Loading the data in batches

```python
#First specify which data quality cuts to make, for most analyses all of these cuts will be needed, 
#by specifying the cuts here the data automatically has cuts applied when it is loaded in in the iterate loop

loader.apply_mPMT_data_quality_cuts() # window/hit-level mPMT quality masks
loader.apply_vme_event_quality_cuts() #vme digitisation and event quality cuts
loader.apply_t5_event_quality_cuts() #T5 selection cuts 

n_windows_passing =0
i_batch =0
for batch in loader.iterate(verbose=False, step_size="100 MB"):
    i_batch+=1
    ##do analysis on the batch here in this example we will just 
    ## check the number of windows that pass the cuts
    n_windows_passing += len(batch)
    # print(f"  Batch: {len(batch)} windows passing cuts")
print("Loaded ",n_windows_passing, "across",i_batch,"batches")
```
